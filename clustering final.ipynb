```python
# Importing the necessary libraries  
import pandas as pd  # For data manipulation  
import numpy as np  # For numerical operations  
import matplotlib.pyplot as plt  # For plotting  
import seaborn as sns  # For enhanced visualizations  
from sklearn.datasets import load_iris  # To load the Iris dataset  
from sklearn.cluster import KMeans  # For KMeans clustering  
from sklearn.cluster import AgglomerativeClustering  # For hierarchical clustering  
from sklearn.decomposition import PCA  # For dimensionality reduction  

# 1. Loading the Iris Dataset  
# Purpose: Load the dataset that we will analyze  
iris = load_iris()  
# X contains the features: measurements of the iris flowers  
X = iris.data  
# y contains the actual species labels (which we’ll ignore for clustering)  
y = iris.target  

# 2. KMeans Clustering  
# Purpose: Identify and visualize natural clusters in the dataset using KMeans  
# Setting up KMeans to find 3 clusters (since we know there are 3 species)  
kmeans = KMeans(n_clusters=3, random_state=42)  
# Fitting the model to our data  
kmeans.fit(X)  
# Predicting the clusters for each flower  
y_kmeans = kmeans.predict(X)  

# Visualizing the KMeans clusters  
# Using PCA to reduce dimensions to 2D for easier plotting  
pca = PCA(n_components=2)  
X_pca = pca.fit_transform(X)  

# Creating a scatter plot for KMeans results  
plt.figure(figsize=(10, 6))  
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k', s=100)  
# Get the cluster centers and transform them to the 2D PCA space  
centers = pca.transform(kmeans.cluster_centers_)  
# Marking the centers on the plot with a red 'X'  
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  
plt.title('KMeans Clustering of Iris Dataset')  
plt.xlabel('PCA Component 1')  
plt.ylabel('PCA Component 2')  
plt.show()  # Display the plot  

# 3. Hierarchical Clustering  
# Purpose: Use a different clustering technique to identify the same groups within the data  
# Setting up the agglomerative clustering to find 3 clusters  
hierarchical = AgglomerativeClustering(n_clusters=3)  
# Fitting the model and predicting clusters  
y_hierarchical = hierarchical.fit_predict(X)  

# Visualizing the Hierarchical clustering results  
plt.figure(figsize=(10, 6))  
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_hierarchical, cmap='viridis', marker='o', edgecolor='k', s=100)  
plt.title('Hierarchical Clustering of Iris Dataset')  
plt.xlabel('PCA Component 1')  
plt.ylabel('PCA Component 2')  
plt.show()  # Display the plot
```
